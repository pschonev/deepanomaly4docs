```python
import pandas as pd

files = ["/home/philipp/projects/dad4td/reports/clustering/0001_cluster_roberta_lowered.tsv"]
df = pd.concat([pd.read_csv(filename, sep="\t") for filename in files]).reset_index(drop=True)

df_mean = df.groupby("hash").mean()
df_mean = df_mean.sort_values("f1_macro")

score_cols = ['completeness',
              'f1_macro',
              'homogeneity',
              'in_f1',
              'out_f1',
              'out_prec',
              'out_rec',
              'v_measure',
             'cluster_n']
data_cols = ['Unnamed: 0',
             'contamination',
             'data_frac',
             'seed']
param_cols = [x for x in list(df.columns) if x not in score_cols]
param_cols = [x for x in param_cols if x not in data_cols]
cols = score_cols + param_cols
df
```

```python
# best f1 macro
df_mean = df.groupby("hash")[score_cols].mean()
df_mean = df_mean.reset_index()
df_mean = df_mean.sort_values(by="homogeneity", ascending=False).reset_index(drop=True)
df_mean.head(20)
```

```python
#worst f1 macro
df_mean = df.groupby("hash")[score_cols].mean()
df_mean = df_mean.reset_index()
df_mean = df_mean.sort_values(by="f1_macro", ascending=True).reset_index(drop=True)
df_mean.head(20)
```

```python
df_mean = df.groupby("mix_ratios_")[score_cols].mean()
df_mean = df_mean.reset_index()
df_mean = df_mean.sort_values(by="f1_macro", ascending=False).reset_index(drop=True)
df_mean.head(20)
```

```python
df_mean = df.groupby("n_comps_")[score_cols].mean()
df_mean = df_mean.reset_index()
df_mean = df_mean.sort_values(by="f1_macro", ascending=False).reset_index(drop=True)
df_mean.head(20)
```
