```python
import pandas as pd

#files = ["/home/philipp/projects/dad4td/reports/clustering/0001_full_doc2vec.tsv", "/home/philipp/projects/dad4td/reports/clustering/0001_doc2vecwikiimdb20news013030.tsv"]
files = ["/home/philipp/projects/dad4td/reports/clustering/0001_longformer_eval.tsv", "/home/philipp/projects/dad4td/reports/clustering/0001_longformer_eval_big_clusters.tsv",
         "/home/philipp/projects/dad4td/reports/clustering/0001_longformer_large_bigger_clusters.tsv",
        "/home/philipp/projects/dad4td/reports/clustering/0001_full_doc2vec.tsv", "/home/philipp/projects/dad4td/reports/clustering/0001_doc2vecwikiimdb20news013030.tsv",
        "/home/philipp/projects/dad4td/reports/clustering/0001_test_transformer.tsv","/home/philipp/projects/dad4td/reports/clustering/0002_test_transformer.tsv"]
df = pd.concat([pd.read_csv(filename, sep="\t") for filename in files]).reset_index(drop=True)

df_mean = df.groupby("hash").mean()
df_mean = df_mean.sort_values("f1_macro")

score_cols = ['completeness',
              'model_train_data',
              'f1_macro',
              'homogeneity',
              'in_f1',
              'out_f1',
              'v_measure',
              'cluster_n',
              'model_data_frac',
              'model_epochs',
              'model_min_count','out_f1_LOF'
             ]
data_cols = ['Unnamed: 0',
             'contamination',
             'data_frac',
             'seed',
              'out_prec',
              'out_rec']
param_cols = [x for x in list(df.columns) if x not in score_cols]
param_cols = [x for x in param_cols if x not in data_cols]
cols = score_cols + param_cols
df
```

```python
# best parameters

# look at v_measure, homogeneity, out_f1 or out_f1_LOF
score = "out_f1"

df_mean = df.groupby(["model_desc", "model_train_data", "n_comps", "mix_ratio", "min_cluster_size"])[score_cols].mean()
df_mean["runs"] = df.groupby(["model_desc", "model_train_data", "n_comps", "mix_ratio", "min_cluster_size"])[score_cols].size()
df_mean = df_mean.reset_index()
df_mean = df_mean[df_mean["runs"]>=3]
df_mean = df_mean.sort_values(by=score, ascending=False).reset_index(drop=True)
df_mean.head(50)
```
